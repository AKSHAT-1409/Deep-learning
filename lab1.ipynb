Write a program to implement AND OR gates using Perceptron.



```python


def perceptron_AND(x1, x2):
    w1 = 1
    w2 = 1
    bias = -1.5
    net = w1*x1 + w2*x2 + bias
    if net >= 0:
        return 1
    else:
        return 0

print("AND Gate")
inputs = [(0,0), (0,1), (1,0), (1,1)]
for x1, x2 in inputs:
    print(f"{x1} AND {x2} = {perceptron_AND(x1, x2)}")
```

```python
def perceptron_OR(x1, x2):
    w1 = 1
    w2 = 1
    bias = -0.5
    net = w1*x1 + w2*x2 + bias
    if net >= 0:
        return 1
    else:
        return 0

print("AND Gate")
inputs = [(0,0), (0,1), (1,0), (1,1)]
for x1, x2 in inputs:
    print(f"{x1} OR {x2} = {perceptron_OR(x1, x2)}")
```

Implementation of XOR Problem using PyTorch Neural Network.

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
X = torch.tensor([[0.,0.],
                  [0.,1.],
                  [1.,0.],
                  [1.,1.]])

Y = torch.tensor([[0.],
                  [1.],
                  [1.],
                  [0.]])

class XORNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(2, 2)
        self.fc2 = nn.Linear(2, 1)

    def forward(self, x):
        x = torch.sigmoid(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x

model = XORNet()
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.1)

for epoch in range(5000):
    optimizer.zero_grad()
    output = model(X)
    loss = criterion(output, Y)
    loss.backward()
    optimizer.step()

print(torch.round(model(X)))
```

5) Implement Simple below Neural Network to solve regression problem.

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python

X = torch.tensor([[1., 2.],
                  [2., 1.],
                  [3., 4.],
                  [4., 3.]], dtype=torch.float32)

Y = torch.tensor([[8.],
                  [7.],
                  [18.],
                  [17.]], dtype=torch.float32)
```

```python
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()

        self.hidden = nn.Linear(2, 2)
        

        self.output = nn.Linear(2, 1)
        
    def forward(self, x):
        h = torch.relu(self.hidden(x))   
        y = self.output(h)               
        return y
```

```python
model = SimpleNN()

criterion = nn.MSELoss()    
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
epochs = 2000

for epoch in range(epochs):
    optimizer.zero_grad()
    
    predictions = model(X)
    loss = criterion(predictions, Y)
    
    loss.backward()
    optimizer.step()
    
    if (epoch + 1) % 500 == 0:
        print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```

```python
with torch.no_grad():
    test_input = torch.tensor([[5., 6.]])
    output = model(test_input)
    print("Predicted output:", output.item())
```
